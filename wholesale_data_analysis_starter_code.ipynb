## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df.loc[:,'line_subtotal'] = df.apply(lambda x: x.qty * x.unit_price, axis=1)
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def calc_shipping(row):
    pplb = 7
    if row.unit_weight * row.qty <= 50:
        pplb = 10
    
    return pplb * row.unit_weight * row.qty

df.loc[:,'shipping_price'] = df.apply(calc_shipping, axis=1)
df
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

df.loc[:,'line_total_price'] = df.apply(lambda x: ( ( (x.line_subtotal+x.shipping_price) *.0925) + (x.line_subtotal+x.shipping_price) ), axis=1)
df
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df.loc[:,'line_cost'] = df.apply(lambda x: (x.qty*x.unit_cost) + x.shipping_price, axis=1)
df
```

```python
# Create a column for the profit of
# each line using line cost and line price

df.loc[:,'line_profit'] = df.apply(lambda x: x.line_total_price - x.line_cost, axis=1)
df
```

Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

#top_5_totals = top_5_clients.loc[:,['line_total_price','line_profit']].groupby(by='order_id').sum()

df.set_index('client_id') \
    .loc[top_5_clients,['line_total_price']] \
    .groupby(by='client_id').sum()['line_total_price'] \
    .apply(format_currency)

# answer:
# 33615: $8,377,308.52
# 66037: $10,259,514.79
# 46820: $9,743,794.36
# 38378: $12,906,550.87
# 24741: $82,268,892.02
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

summary_df = df.set_index('client_id') \
    .loc[top_5_clients,['qty','shipping_price','line_total_price','line_profit']] \
    .groupby(by='client_id') \
    .apply(lambda x: pd.Series( {'Total Quantity': x['qty'].sum() 
                                , "Total Shipping Price": pd.Series(x['shipping_price'].sum()).apply(format_currency_M).sum()
                                , "Total Price": pd.Series(x['line_total_price'].sum()).apply(format_currency_M).sum()
                                , "Total Profit": pd.Series(x['line_profit'].sum()).apply(format_currency_M).sum() })) \
                                .sort_values('Total Profit')
    

#summary_df['line_total_price'] = summary_df['line_total_price'].apply(format_currency)

summary_df
```
