## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

#top_client.loc[:,'line_subtotal'] = top_client.apply(lambda x: x.qty * x.unit_price, axis=1)
#top_client
df.loc[:,'line_subtotal'] = df.apply(lambda x: x.qty * x.unit_price, axis=1)
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def calc_shipping(row):
    pplb = 7
    if row.unit_weight <= 50:
        pplb = 10
    
    return pplb * row.unit_weight


# top_client.loc[:,'shipping_price'] = top_client.apply(calc_shipping, axis=1)
# top_client

df.loc[:,'shipping_price'] = df.apply(calc_shipping, axis=1)
df
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# top_client.loc[:,'line_total_price'] = top_client.apply(lambda x: x.line_subtotal + x.shipping_price + (x.line_subtotal*.0925), axis=1)
# top_client

df.loc[:,'line_total_price'] = df.apply(lambda x: ( (x.line_subtotal + x.shipping_price) *1.0925), axis=1)
df
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

# top_client.loc[:,'line_cost'] = top_client.apply(lambda x: (x.qty*x.unit_cost) + x.shipping_price, axis=1)

# top_client

df.loc[:,'line_cost'] = df.apply(lambda x: (x.qty*x.unit_cost) + x.shipping_price, axis=1)
df
```

```python
# Create a column for the profit of
# each line using line cost and line price

# top_client.loc[:,'line_profit'] = top_client.apply(lambda x: x.line_total_price - x.line_cost, axis=1)
# top_client

df.loc[:,'line_profit'] = df.apply(lambda x: x.line_total_price - x.line_cost, axis=1)
df
```
